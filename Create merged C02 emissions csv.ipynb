```python
import pandas as pd
```

```python
### Pulled this data from https://www.climatetrace.org/inventory in 02/2022
#### Disclaimer from the source that should be noted: "All Sectors excluding Forest and Land Use Emissions"
df = pd.read_csv("climatetrace_emissions_by_subsector_timeseries_all.csv")
df
```

```python
##Manipulations for climate trace data cleaning
df = df.dropna()
df = df.rename(columns = {'Tonnes Co2e':'Tonnes_C02'})
df['Tonnes_C02_Millions'] = df['Tonnes_C02']/1000000
df['year'] = pd.DatetimeIndex(df['start']).year
df = df.rename(columns = {'country':'country_code'})
df.year = df.year.astype(str)
df
```

```python
### Pulling in population data and country categories from https://data.worldbank.org/indicator/SP.POP.TOTL in 02/2022
wpop = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_3628828.csv",skiprows=4)
wpop
```

```python
##Changes to the population dataframe so I can easily merge it with the other dataframe
wpop = wpop.loc[:, ~wpop.columns.str.contains('^Unnamed')]
wpop = wpop.drop(columns=['Indicator Name','Indicator Code'])
wpop = wpop.rename(columns={'Country Code':'country_code'})
keepcolumns = ['Country Name','country_code']
value_columns = wpop.columns.difference(keepcolumns)
wpop = pd.melt(wpop,id_vars=keepcolumns,value_vars=value_columns,var_name='Year',
        value_name='Population')
wpop
```

```python
###Merging the two dataframes and a quick per capita calculation
dfp = pd.merge(df,wpop,how='left',left_on=['country_code','year'],right_on=['country_code','Year'])
dfp.reset_index(drop=True,inplace=True)
dfp['Pounds_C02_PerCapita'] = dfp['Tonnes_C02']/dfp['Population']*2000
dfp = dfp.drop(columns=['Year','Country Name'])
dfp = dfp.rename(columns={'country_full':'Country_Name'})
dfp
```

```python
###Bringing in Groupings from the world bank to category and remove outliers
wg = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_3628828.csv')
wg = wg[['Country Code','Region','IncomeGroup']]
dfm = pd.merge(dfp,wg,how='left',left_on='country_code',right_on='Country Code')
dfm = dfm.drop(columns=['Country Code'])
dfm
```
